# Configuração das bibliotecas necessárias

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

# Armazenamento dos dados
Os dados do banco de dados são armazenados em um DataFrame da biblioteca Pandas

```python
solda1_df = pd.read_sql('SELECT * FROM aquisicao_solda', engine)
```

# Teste de normalidade Anderson-Darling

```python
t1_norm = stats.anderson(solda1_df['tensao_1'], dist="norm")

if t1_norm.statistic < 0.05:
    print('Os dados seguem uma distribuição normal')
else:
    print('Os dados não seguem uma distribuição normal')
    
t2_norm = stats.anderson(solda1_df['tensao_2'], dist="norm")

if t2_norm.statistic < 0.05:
    print('Os dados seguem uma distribuição normal')
else:
    print('Os dados não seguem uma distribuição normal')
    
c1_norm = stats.anderson(solda1_df['corrente_1'], dist="norm")

if c1_norm.statistic < 0.05:
    print('Os dados seguem uma distribuição normal')
else:
    print('Os dados não seguem uma distribuição normal')
    
c2_norm = stats.anderson(solda1_df['corrente_2'], dist="norm")

if c2_norm.statistic < 0.05:
    print('Os dados seguem uma distribuição normal')
else:
    print('Os dados não seguem uma distribuição normal')
    
v1_norm = stats.anderson(solda1_df['varame_1'], dist="norm")

if v1_norm.statistic < 0.05:
    print('Os dados seguem uma distribuição normal')
else:
    print('Os dados não seguem uma distribuição normal')
```

# Estatísticas descritivas

```python
# Níveis de Índice
outside = ['Fonte 1','Fonte 1','Fonte 1','Fonte 2','Fonte 2','Fonte 2']
inside = ['Tensão', 'Corrente', 'Velocidade de Arame', 'Tensão', 'Corrente', 'Velocidade de Arame']
hier_index = pd.MultiIndex.from_tuples(list(zip(outside,inside))) # zip: une as listas, elemento por elemento
estat = pd.DataFrame([[solda1_df['tensao_1'].mean(), solda1_df['tensao_1'].std()], [solda1_df['corrente_1'].mean(), solda1_df['corrente_1'].std()], [solda1_df['varame_1'].mean(), solda1_df['varame_1'].std()], [solda1_df['tensao_2'].mean(), solda1_df['tensao_2'].std()], [solda1_df['corrente_2'].mean(), solda1_df['corrente_2'].std()], [solda1_df['varame_2'].mean(), solda1_df['varame_2'].std()]], index = hier_index, columns=['Média', 'Desvio-Padrão'])
estat
```
