<h3>Sectional area</h3>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Input: Dimensões da junta
print('---Dimensões da Junta---')
deg = float(input('Insira o valor do ângulo da junta: '))
nose = float(input('Insira o valor do nariz da junta: '))
gap = float(input('Insira o valor do gap da junta: '))
thick = float(input('Insira o valor da espessura da chapa: '))

# Cálculos da área da secção transversal
area = ((np.tan(np.radians(deg/2)))*(thick-nose)**2) + (thick*gap)
print('-------')
print(f'A área frontal da junta é {area:.2f} mm²\n')
```

<h3>Evaluation of geometrical parameters influence on the joint's sectional area</h3>

<p>
    Para a avaliação da influência do valor do nariz, serão fixados os seguintes parâmetros:
</p>
<ul>
    <li>Ângulo da junta = 45°</li>
    <li>Abertura de raiz = 2,5 mm</li>
    <li>Espessura da chapa = 25,4 mm</li>
</ul>

<hr>

<p>
    Para a avaliação da influência do valor da abertura de raiz, serão fixados os seguintes parâmetros:
</p>
<ul>
    <li>Ângulo da junta = 45°</li>
    <li>Nariz = 2,0 mm</li>
    <li>Espessura da chapa = 25,4 mm</li>
</ul>

<hr>

<p>
    Para a avaliação da influência do valor da espessura da junta, serão fixados os seguintes parâmetros:
</p>
<ul>
    <li>Ângulo da junta = 45°</li>
    <li>Abertura de raiz = 2,5 mm</li>
    <li>Nariz = 2,0 mm</li>
</ul>

<hr>

<p>
    Para a avaliação da influência do valor do ângulo da junta, serão fixados os seguintes parâmetros:
</p>
<ul>
    <li>Nariz = 2,0 mm</li>
    <li>Abertura de raiz = 2,5 mm</li>
    <li>Espessura da chapa = 25,4 mm</li>
</ul>

```python
# Parâmetros fixados
deg = 45
gap = 2.5
thick = 25.4
nose = 2.0

# Nariz
nose1 = np.linspace(1.0, 5.0, 100)
area1 = ((np.tan(np.radians(deg/2)))*(thick-nose1)**2) + (thick*gap)

# Abertura de raiz
gap1 = np.linspace(1.0, 5.0, 100)
area2 = ((np.tan(np.radians(deg/2)))*(thick-nose)**2) + (thick*gap1)

# Espessura da junta
thick1 = np.linspace(6.4, 50.8, 100)
area3 = ((np.tan(np.radians(deg/2)))*(thick1-nose)**2) + (thick*gap)

# Ângulo da junta
deg1 = np.linspace(10, 75, 100)
area4 = ((np.tan(np.radians(deg1/2)))*(thick1-nose)**2) + (thick*gap)

# Plotagem das curvas
font = {"family": "serif",
        "size": 12,
        }

sns.set_style('whitegrid')

fig, axes = plt.subplots(2, 2, sharey=False, sharex=False, figsize=(1280/96, 720/96))
fig.suptitle('Influência da Geometria da Junta na Área Transversal', fontdict=font)

sns.lineplot(ax=axes[0,0], x=nose1, y=area1)
axes[0,0].set_ylabel('Área [mm²]', fontdict=font)
axes[0,0].set_xlabel('Nariz [mm]', fontdict=font)

sns.lineplot(ax=axes[0,1], x=gap1, y=area2)
axes[0,1].set_xlabel('Abertura de raiz [mm]', fontdict=font)

sns.lineplot(ax=axes[1,0], x=thick1, y=area3)
axes[1,0].set_ylabel('Área [mm²]', fontdict=font)
axes[1,0].set_xlabel('Espessura da junta [mm]', fontdict=font)

sns.lineplot(ax=axes[1,1], x=deg1, y=area4)
axes[1,1].set_xlabel('Ângulo [°]', fontdict=font)

plt.tight_layout()
plt.show()

fig.savefig('Avaliação da Influência.png', dpi=fig.dpi)
```

<h2>Response Surfaces</h2>

<h3>Libraries</h3>

```python
from sklearn import linear_model
import statsmodels.api as sm
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Colormap
colormap = 'jet'
```

```python
# Nariz
nose = np.linspace(1.0, 5.0, 100)

# Abertura de raiz
gap = np.linspace(1.0, 5.0, 100)

# Espessura da junta
thick = np.linspace(6.4, 50.8, 100)

# Ângulo da junta
deg = np.linspace(10, 75, 100)

# Área transversal
area = ((np.tan(np.radians(deg/2)))*(thick-nose)**2) + (thick*gap)

data = {
    'Nose': nose,
    'Gap': gap,
    'Thickness': thick,
    'Angle': deg,
    'Area': area
}

df = pd.DataFrame(data, columns=['Nose', 'Gap', 'Thickness', 'Angle', 'Area'])
print(df)

X2, Y2 = np.meshgrid(df['Gap'], df['Thickness'])
Z2 = ((np.tan(np.radians(45/2)))*(Y2-2)**2) + (Y2*X2)

fig = plt.figure(figsize=(12.8, 9.6))

ax = plt.axes(projection='3d')
p = ax.plot_surface(X2, Y2, Z2, cmap=colormap, linewidth=0, alpha=0.8)
ax.contour(X2, Y2, Z2, zdir='Y2', offset=-5, cmap=colormap)

ax.set_xlabel('Abertura de raiz [mm]', fontdict=font)
ax.set_ylabel('Espessura da junta [mm]', fontdict=font)
ax.set_zlabel('Área transversal [mm²]', fontdict=font)

ax.set_title('Superfície de Resposta - Abertura de raiz X Espessura da junta', fontdict=font)
fig.colorbar(p, ax=ax, shrink=0.75)
```

```python
X3, Y3 = np.meshgrid(df['Angle'], df['Nose'])
Z3 = ((np.tan(np.radians(45/2)))*(Y3-2)**2) + (Y3*X3)

fig = plt.figure(figsize=(12.8, 9.6))

ax = plt.axes(projection='3d')
q = ax.plot_surface(X3, Y3, Z3, cmap=colormap, linewidth=0, alpha=0.8)
ax.contour(X3, Y3, Z3, zdir='Y3', offset=-5, cmap=colormap)

ax.set_xlabel('Ângulo da junta [°]', fontdict=font)
ax.set_ylabel('nariz da junta [mm]', fontdict=font)
ax.set_zlabel('Área transversal [mm²]', fontdict=font)

ax.set_title('Superfície de Resposta - Ângulo da junta X Nariz da junta', fontdict=font)
fig.colorbar(q, ax=ax, shrink=0.75)
```
